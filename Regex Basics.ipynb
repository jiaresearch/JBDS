### Regex Basics
Use Regex for pattern matching in texts
- to check if a pattern exist
- to get all instances of a particular pattern
- to split strings

```python
import re
```

```python
text = "This is a good day."

# the re.search returns a match project (this is considered to pass the "if" check)
if re.search("good", text):
    print('Matched')

if re.search("come", text):
    print('Matched')
else:
    print('Not Matched')
```
