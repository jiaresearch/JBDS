# Pandas

```python
import pandas as pd
```

## Pandas Series

```python
# Create series: pass in a list
students = ['Alice', 'Jack', 'Molly']
print(pd.Series(students))
```

```python
numbers = [1, 2, 3]
print(pd.Series(numbers))
```

```python
# How does pandas handle missing data for str dtype series?
# Starting from pandas 3.0, str type is enabled by default
students = ['Alice', 'Jack', None]
print(pd.Series(students))
```

```python
# How does pandas handle missing data for numerical series?
numbers = [1, 2, None]
print(pd.Series(numbers))
```
