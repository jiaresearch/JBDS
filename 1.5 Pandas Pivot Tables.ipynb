# Pandas Pivot Tables

```python
import pandas as pd
import numpy as np
```

A pivot table is itself a DataFrame, where the rows represent one variable that you're interested in, the columns another, and the cell's some aggregate value. A pivot table also tends to includes marginal values as well, which are the sums for each column and row.

```python
# Load the Times Higher Education World University Ranking dataset
df = pd.read_csv('data_intro/cwurData.csv')
df.head()
```

```python
# We are interested in creating a tier level based on world ranks
def create_category(ranking):
    # Since the rank is just an integer, I'll just do a bunch of if/elif statements
    if (ranking >= 1) & (ranking <= 100):
        return "First Tier Top Unversity"
    elif (ranking >= 101) & (ranking <= 200):
        return "Second Tier Top Unversity"
    elif (ranking >= 201) & (ranking <= 300):
        return "Third Tier Top Unversity"
    return "Other Top Unversity"

# Apply this to the world_rank column
df['Rank_Level'] = df['world_rank'].apply(lambda x: create_category(x))
df1 = df[['Rank_Level', 'institution', 'country', 'score', 'year']]
df1.head()
```

```python
# Use a pivot_table to find that for each country, what are the average scores for each tier
df1.pivot_table(values='score', index='country', columns='Rank_Level',
                aggfunc=[np.mean]).head()
```

```python
# Find that for each country, what are the average and max scores for each tier
df1.pivot_table(values='score', index='country', columns='Rank_Level',
                aggfunc=[np.mean, np.max]).head()
```

```python
# Add margins (i.e., total averages, and total max)
df2 = df1.pivot_table(values='score', index='country', columns='Rank_Level',
                      aggfunc=[np.mean, np.max],
                      margins=True)
df2.head()
```

```python
# access the index and columns of a pivot table (which is essentially multi-level dataframe)
print(df2.index)
print("----------------------------")
print(df2.columns)
```

```python
# Query pivot table with chained reference
# The following query returns a series
df2['mean']['First Tier Top Unversity'].head()
```

```python
# Find which country has the maximum average score on first tier university level?
df2['mean']['First Tier Top Unversity'].idxmax()
```

```python
# We can use stacking to move the multi-layer index to the row indices
df3=df2.stack()
df3.head(10)
```

```python
# we can also unstack to return to the previous df
df3.unstack().head()
```

```python
# if unstack again:
df3.unstack().unstack()
```
