# Pandas Series and DataFrames

```python
import pandas as pd
import numpy as np
```

## Pandas Series

### Create Pandas Series

```python
# Create series: pass in a list
students = ['Alice', 'Jack', 'Molly']
print(pd.Series(students))
```

```python
numbers = [1, 2, 3]
print(pd.Series(numbers))
```

### Missing values in Pandas Series

```python
# How does pandas handle missing data for str dtype series?
# Starting from pandas 3.0, str type is enabled by default
students = ['Alice', 'Jack', None]
print(pd.Series(students))
```

```python
# How does pandas handle missing data for numerical series?
numbers = [1, 2, None]
print(pd.Series(numbers))
```

```python
# Equality checks on missing values np.nan do not work
print(np.nan == np.nan)
print(np.nan == None)
print(None == None)
```

```python
# Instead, use isnan() to test missing values
print(np.isnan(np.nan))
```

### Create Pandas Series with Index

```python
students_scores = {'Alice': 'Physics',
                   'Jack': 'Chemistry',
                   'Molly': 'English'}
s = pd.Series(students_scores)
print(s)
print(s.index)
```

```python
# Separate index creation from the data
s = pd.Series(['Physics', 'Chemistry', 'English'], index=['Alice', 'Jack', 'Molly'])
print(s)
print(s.index)
```

```python
# When index do not align with the dictionary keys
# The unaligned index will contain a missing value
students_scores = {'Alice': 'Physics',
                   'Jack': 'Chemistry',
                   'Molly': 'English'}
s = pd.Series(students_scores, index=['Alice', 'Molly', 'Sam'])
print(s)
```

### Query Pandas Series

```python
students_classes = {'Alice': 'Physics',
                    'Jack': 'Chemistry',
                    'Molly': 'English',
                    'Sam': 'History'}
s = pd.Series(students_classes)
s
```

```python
# iloc and loc: they are attributes, not methods. So use square brackets, not parentheses.

# use iloc to query the value of the forth entry
print(s.iloc[3])

# use iloc to query the value to the key 'Molly'
print(s.loc['Molly'])

# Pandas will behave as if iloc is used, when an integer parameter is passed in: e.g. s[3], s['Molly']
print(s['Molly'])

# However, Pandas can't determine automatically whether you're intending to query
# by index poistion or index label, if your index label is numeric.
# So the safer option is to be more explicit and use iloc or loc attributes directly.
```

```python
# Two methods to aggregate a series
grades = pd.Series([90, 80, 70, 60])

# method 1: iterate over the series, sum, and divide
total = 0
for grade in grades:
    total += grade
print(total/len(grades))

# method 2: use numpy sum()
# Recommended: the numpy function utilizes parallel computing features.
import numpy as np
total = np.sum(grades)
print(total/len(grades))
```

```python
numbers = pd.Series(np.random.randint(0, 1000,10000))
```

```python
# Related feature: broadcasting - apply operations to every value in the series.
# e.g., we could use += operator directly on the Series object
numbers = pd.Series(np.random.randint(0, 1000,10000))
print(numbers.head())

numbers += 2
print(numbers.head())
```

```python
# The .loc attribute lets you not only to modify data in place, but also add new data as well.
# If the value you pass to index doesn't exist, then a new entry is added.
s = pd.Series([1, 2, 3])

# let's add some new value
s.loc['History'] = 102
print(s)
```

```python
# Pandas with not unique index values
students_classes = pd.Series({'Alice': 'Physics',
                    'Jack': 'Chemistry',
                    'Molly': 'English',
                    'Sam': 'History'})
print(students_classes)
print()

kelly_classes = pd.Series(['Philosophy', 'Arts', 'Math'], index=['Kelly', 'Kelly', 'Kelly'])
print(kelly_classes)
```

```python
# Concat the two series
all_students_classes = pd.concat([students_classes, kelly_classes])
all_students_classes
```

## Dataframes

### Creating Dataframes

```python
import pandas as pd
# method 1 to create a dataframe: list of series (each series is a row)
record1 = pd.Series({'Name':'Alice',
                     'Class':'Physics',
                     'Score': 85})
record2 = pd.Series({'Name':'Jack',
                     'Class':'Chemistry',
                     'Score': 82})
record3 = pd.Series({'Name':'Helen',
                     'Class':'Biology',
                     'Score': 90})
df = pd.DataFrame([record1, record2, record3],
                  index=['school1', 'school2', 'school1'])
df.head()
```

```python
# Method 2 to create a dataframe: list of dictionaries (each is a row)
students = [{'Name':'Alice',
             'Class':'Physics',
             'Score': 85},
            {'Name':'Jack',
             'Class':'Chemistry',
             'Score': 82},
            {'Name':'Helen',
             'Class':'Biology',
             'Score': 90}]
df = pd.DataFrame(students, index=['school1', 'school2', 'school1'])
print(df)
```

### Query the dataframe rows and columns

```python
# query by index name
df.loc['school2']
```

```python
# Since school1 correspondes to multiple records, the .loc will return a new dataframe
df.loc['school1']
```

```python
# select based on row and columns
df.loc['school1', 'Name']
```

```python
# to get a column - method 1: transpose
df.T.loc['Name']
```

```python
# to get a column - method 2: df[]
print(df['Name'])

# a single column  projection is a Series object
print(type(df['Name']))
```

```python
# select by both rows and columns - method 1: chaining
df.loc['school1']['Name']
```

```python
# select by both rows and columns - method 2: use .loc
df.loc[:, ['Name', 'Score']]
```

### Sort Values

```python
df.sort_values(by=['Score'], ascending=True)
```

### Dropping data

```python
# Dropping data without overwriting - drop a row by index
print(df.drop('school1'))
print("---------------------------------------------------")

# Dropping data without overwriting - drop a column
# axis = 1 means dropping a column
print(df.drop('Name', axis=1))
print("---------------------------------------------------")

# Dropping data with overwriting
copy_df = df.copy()
copy_df.drop('Name', inplace=True, axis=1)
print(copy_df)
print("---------------------------------------------------")

# Dropping data with overwriting - use del
copy_df = df.copy()
del copy_df['Class']
print(copy_df)
```
