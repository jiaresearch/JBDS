# Pandas Date and Time

Pandas has four main time related classes. **Timestamp**, **DatetimeIndex**, **Period**, and **PeriodIndex**.
- **Timestamp**: represents a single timestamp and associates values with points in time
- **DatetimeIndex**:  The index of a timestamp is DatetimeIndex
- **Period**: represents a single time span, such as a specific day or month.
- **PeriodIndex**: The index of a Period is PeriodIndex

```python
import pandas as pd
import numpy as np
```

## Timestamp

```python
# Timestamp is interchangeable with Python's datetime in most cases.
pd.Timestamp('9/1/2019 10:05AM')
```

```python
# Create a timestamp by passing in year, month, date, hour, minute, second
timestp = pd.Timestamp(2026, 12, 20, 18, 6, 30)
timestp
```

```python
# show the weekday of the timestamp, 1 represents Monday and 7 represents Sunday
timestp.isoweekday()
```

```python
# extract year, month, day, hour, minute, second from a timestamp
print(timestp.year)
print(timestp.hour)
print(timestp.second)
```

## Period

```python
# Create a period that is January 2016,
pd.Period('1/2026')
```

```python
# Create a period that is Mar 5, 2016
pd.Period('3/5/2026')
```

```python
# Perform subtraction (i.e., back by 2 days)
pd.Period('3/5/2026') - 2
```

## DatetimeIndex and PeriodIndex

```python
# Create a Series with timestamp index --> a DatetimeIndex
t1 = pd.Series(list('abc'), index = [pd.Timestamp('2016-09-01'),
                                     pd.Timestamp('2016-09-02'), pd.Timestamp('2016-09-03')])
t1
```

```python
type(t1.index)
```

```python
# Similarly, we can create a period-based index as well.
t2 = pd.Series(list('def'), [pd.Period('2016-09'), pd.Period('2016-10'),
                             pd.Period('2016-11')])
t2
```

```python
type(t2.index)
```

## Converting to Datetime

```python
# Now, let's look into how to convert to Datetime. Suppose we have a list of dates as strings and we want to
# create a new dataframe

# I'm going to try a bunch of different date formats
d1 = ['2 June 2013', 'Aug 29, 2014', '2015-06-26', '7/12/16']

# And just some random data
ts3 = pd.DataFrame(np.random.randint(10, 100, (4,2)), index=d1,
                   columns=list('ab'))
ts3
```

```python
# Using pandas to_datetime, pandas will try to convert these to Datetime and put them in a standard format.

ts3.index = pd.to_datetime(ts3.index, format='mixed')
ts3
```

```python
# to_datetime also() has options to change the date parse order. For example, we
# can pass in the argument dayfirst = True to parse the date in European date.

pd.to_datetime('4.7.12', dayfirst=True)
```

## Timedelta (time differences)

```python
# The difference between September 3rd and September 1st -> 2 days
pd.Timestamp('9/3/2016')-pd.Timestamp('9/1/2016')
```

```python
# We can also do something like find what the date and time is for
# 12 days and three hours past September 2nd, at 8:10 AM.
pd.Timestamp('9/2/2026 8:10AM') + pd.Timedelta('12D 3h')
```

## Offset

Offset is similar to timedelta, but it follows specific calendar duration rules. Offset allows flexibility in terms of types of time intervals. Besides hour, day, week, month, etc it also has business day, end of month, semi month begin etc

```python
# Let's create a timestamp, and see what day is that
pd.Timestamp('9/4/2016').weekday()
```

```python
# Now we can now add the timestamp with a week ahead
pd.Timestamp('9/4/2016') + pd.offsets.Week()
```

```python
# Now let's try to do the month end, then we would have the last day of Septemer
pd.Timestamp('9/4/2016') + pd.offsets.MonthEnd()
```

## Working with Dates in a Dataframe

```python
# create an index of 9 dates, taken bi-weekly on every Sunday
pd.date_range('10-01-2026', periods=9, freq='2W-SUN')
```

```python
# create an index of 9 consecutive business days.
pd.date_range('10-01-2026', periods=9, freq='B')
```

```python
# create an index of 12 days, increment by quarter, start in June (i.e., June 1)
pd.date_range('04-01-2026', periods=12, freq='QS-JUN')
```

```python
# Now, let's go back to our weekly on Sunday example and create a DataFrame using these dates, and some random data, and see what we can do with it.
dates = pd.date_range('10-01-2026', periods=9, freq='2W-SUN')
df = pd.DataFrame({'Count 1': 100 + np.random.randint(-5, 10, 9).cumsum(),
                  'Count 2': 120 + np.random.randint(-5, 10, 9)}, index=dates)
df
```

```python
# check what day of the week a specific date is
# Note, Monday is 0, Sunday is 6.
df.index.weekday
```

```python
# We can also use diff() to find the difference between each date's value.
df.diff()
```

```python
# Suppose we want to know what the mean count is for each month in our DataFrame. We can do this using resample. Converting from a higher frequency from a lower frequency is called downsampling (we'll talk about this in a moment)
df.resample('ME').mean()
```

```python
# Now let's talk about datetime indexing and slicing, which is a wonderful feature of the pandas DataFrame.
print(df[df.index.year==2027])
print('------------------------------')
# A quicker way is to use partial string indexing (.loc[] is required in recent Pandas versions)
print(df.loc['2027'])
```

```python
# partial indexing with year and month
print(df.loc['2026-12'])
```

```python
# slicing to get a range of dates
print(df.loc['2026-12':])
```
